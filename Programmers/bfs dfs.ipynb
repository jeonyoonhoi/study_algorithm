### 단어변환

- 가장 먼저 할 일은 target이  words 안에 있는가.

- 그 다음 그 인접행렬 또는 리스트의 최단거리를 찾으면 된다.

```python
tickets = [['ICN', 'SFO'], ['ICN', 'ATL'], ['SFO', 'ATL'], ['ATL', 'ICN'], ['ATL','SFO']]
```

```python
answer = []
V='ICN'
visited = [0 for _ in range(len(tickets))]

def findpath(V):
    global answer
    global visited
    global tickets
    
    # 다음에 갈 수 있는 후보군을 찾는다. 
    buffer = []
    for i, x in enumerate(tickets) : 
        if visited[i]==0 and x[0]==V: # 사용하지 않은 티켓이고, 시작점이 현재위치 V 일때 
            buffer.append(i)          # 다음 후보에 추가한다. 

            
    if buffer ==[] :# 다음에 갈 곳이 없으면 나가야지 
        return
        
    minalpha = 'ZZZ'                  # 후보들 중에서는 가장 작은 (알파벳 순서 앞에서부터)  도착지로 가게된다. 
    for y in buffer: # buffer 에는 예비 후보티켓의 인덱스가 담기게 된다. 
        if (tickets[y][1])<minalpha :  
            minalpha = tickets[y][1]
            print(minalpha)
            minindex = y              # 다음에 갈 곳의 티켓의 인덱스
    # minindex 가 가리키는 곳이 다음 노드가 된다. 
    answer.append(V)        # 다음 패스는 이렇게 된다. 
    visited[minindex] = 1   #다음행선지로 갈 곳이 정해졌으므로 티켓을 사용 완료 하였다. 
    findpath(minalpha)      # 다음 살펴볼 노드는 V =minalpha 
    
    
```
